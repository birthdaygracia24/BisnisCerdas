# Melakukan Import Library yang dibutuhkan beserta inisiasi variabel

pandas merupakan library yang digunakan untuk pemrosesan data analisis

numpy merupakan library yang digunakan untuk pemrosesan numerical computing


scikit-learn(sklearn) merupakan library yang digunakan untuk pemrosesan machine learning

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

FEATURE_NAMES = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

iris = pd.read_csv(r'D:\perkuliahan\Semester 7\Bisnis Cerdas\Tugas 3\iris.csv')
X = pd.DataFrame(iris, columns = FEATURE_NAMES)

variant = preprocessing.LabelEncoder()
variant.fit(iris.variety)
y = variant.transform(iris.variety)
```

# Menampilkan variabel X dalam bentuk dataframe

Penyajian data dalam bentuk dataframe dengan menampilkan 5 data teratas dan 5 data terbawah

```python
X[:150]
```

# Menampilkan Data label yang disimpan dalam variabel y

Data yang ditampilkan berupa data yang disimpan dalam array

```python
y[:150]
```

# Pembentukan bagian training dan testing algoritma

Data Testing yang digunakan sebesar 35% dengan nilai dari variabel random_state(bilangan acak yang diberikan) sebesar 6218012 (berdasarkan NIM H06218012)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state = 6218012)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape, y_test.shape)
```

# Import (DecisionTreeClassifier) terhadap library sklearn.tree

```python
from sklearn.tree import DecisionTreeClassifier
```

# Hyperparameter Tuning

Setelah import berhasil dilakukan, proses yang dilakukan adalah Hyperparameter Tuning (Proses pencarian model terbaik dengan melakukan perubahan parameter model sebelum proses training.

Parameter model yang dapat diubah adalah parameter 'criterion' (Penyimpanan metode dalam penghitungan pemisahan dengan metrik yang berbeda) dan 'max_depth' (Penentuan kedalaman tree).

```python
model = DecisionTreeClassifier(criterion="entropy", max_depth = 3)
model.fit(X_train,y_train)
```

```python
model1 = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
model1.fit(X_train,y_train)
```

```python
model2 = DecisionTreeClassifier(criterion="entropy", max_depth = 5)
model2.fit(X_train,y_train)
```

```python
model3 = DecisionTreeClassifier(criterion="entropy", max_depth = 7)
model3.fit(X_train,y_train)
```

```python
modelA = DecisionTreeClassifier(criterion="gini", max_depth = 3)
modelA.fit(X_train,y_train)
```

```python
modelB = DecisionTreeClassifier(criterion="gini", max_depth = 4)
modelB.fit(X_train,y_train)
```

```python
modelC = DecisionTreeClassifier(criterion="gini", max_depth = 5)
modelC.fit(X_train,y_train)
```

```python
modelD = DecisionTreeClassifier(criterion="gini", max_depth = 7)
modelD.fit(X_train,y_train)
```

# Import (export_graphviz) terhadap library sklearn.tree

Import dilakukan untuk melakukan visualisasi pada tree

```python
from sklearn.tree import export_graphviz
export_graphviz(model, 'tree.dot', feature_names = FEATURE_NAMES)
export_graphviz(model1, 'tree1.dot', feature_names = FEATURE_NAMES)
export_graphviz(model2, 'tree2.dot', feature_names = FEATURE_NAMES)
export_graphviz(model3, 'tree3.dot', feature_names = FEATURE_NAMES)
export_graphviz(modelA, 'treeA.dot', feature_names = FEATURE_NAMES)
export_graphviz(modelB, 'treeB.dot', feature_names = FEATURE_NAMES)
export_graphviz(modelC, 'treeC.dot', feature_names = FEATURE_NAMES)
export_graphviz(modelD, 'treeD.dot', feature_names = FEATURE_NAMES)
```
